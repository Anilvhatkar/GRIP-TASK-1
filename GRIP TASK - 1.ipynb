IMPORTING THE REQUIRED LIBRARIES.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

IMPORTING AND READING DATASET.

```python
Url = "http://bit.ly/w-data"
df = pd.read_csv(Url)
print("data imported")
```

```python
#check if there is any missing value in the data
df.isnull==True
```

DATA VISUALIZATION

```python
df.plot(x = "Hours" , y = "Scores" , style = "o")
plt.title( "HOURS V/S PERCENTAGE" )
plt.xlabel( "Hours Studied")
plt.ylabel( "Percentage Scored")
plt.show()
```

SPLIT THE DATASET FOR TRAINING.

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
print("Success")
```

PLOTTNG THE REGRESSION LINE

```python
line = reg.coef_*x + reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x, line, color="red")
plt.title("Hours V/S Percentage")
plt.xlabel("Hours studied")
plt.ylabel("Percentage Scored")
plt.show()
```

**PREDICTING THE TEST SET RESULTS**

```python
print(x_test)                     #Test data : in hours
y_pred = reg.predict(x_test)      #Predicting scores on the basis of test data
print(y_pred)
```

```python
reg.score(x_test, y_test)
print("The R-square of the model is", reg.score(x_test, y_test))
```

**VISULALIZING THE TRAINING TEST RESULTS**

```python
plt.scatter(x_train, y_train, color = "green")
plt.plot(x_train, reg.predict(x_train), c="r")
plt.title("Hours V/S Percentage\n Training set results")
plt.xlabel("Hours studied")
plt.ylabel("Percentage Scored")
plt.show()
```

**VISUALIZING TEST RESULTS**

```python
plt.scatter(x_test, y_test, color="red")
plt.plot(x_train, reg.predict(x_train), color="blue")
plt.title("Hours V/S Percentage\n Test set results")
plt.xlabel("Hours studied")
plt.ylabel("Percentage Scored")
plt.show()
```

**PREDICTING THE PERCENTAGE OF MARKS SCORED**

```python
#Comparing Actual vs Predicted values
df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
df
```

**ERROR METRIC**

```python
from sklearn import metrics
print("Mean absolute error", metrics.mean_absolute_error(y_test, y_pred))
```
